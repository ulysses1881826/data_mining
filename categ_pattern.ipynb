```python
import numpy as np
import pandas as pd
```

```python
categ = open('categories.txt','r')
```

```python
categ.close()
```

```python
data=[]
for i in categ:
    data.append(i)
```

```python
len(data)
```

```python
fre_one = {}
```

```python
for i in range(len(data)):
    data[i] = data[i][0:-1][0:-1]
```

```python
data[-1] = data[-1]+'od'
```

```python
for items in data:
    b = items.split(';')
    for item in b:
        if item not in fre_one:
            fre_one[item] = 0
        else:
            fre_one[item] += 1
```

```python
fre_one
```

```python
patterns = []
one_fre = []
for name,num in fre_one.items():
    if num>771:
        one_fre.append(name)
        patterns.append([name,num])
```

```python
patterns
```

```python
fw = open('patterns.txt','w')
for pattern in patterns:
    str_tmp = str(pattern[1])+':'+pattern[0]
    fw.writelines(str_tmp)
    fw.writelines('\n')
fw.close()
```
